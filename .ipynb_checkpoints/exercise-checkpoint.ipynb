```python
df_sample =  spark.sql(
"SELECT \
    managers.EmployeeId as 'ManagerId',\
    managers.LastName as 'ManagerLastName',\
    employees.lastName as 'ManagerFirstName',\
    employees.ReportTo as 'StaffId' \
    From employees as managers \
    Left join employees on employees.ReportsTo = managers.EmployeeId \
    Group by manager.EmployeeId  \
    order by 3 desc")
```

who are the top 10 customers
--limit--

```python
df_sample4 =  spark.sql(
"SELECT\
    customers.customerId,\
    sum(customers.PostalCode) as TotalSales\
    FROM customers\
    join invoices on invoices.CustomerId = customers.CustomerId\
    group by customers.CustomerId, customers.State \
    order by customers.customerId desc\
    limit 10")
df_sample4.show(10)
```

calculate sales

```python
df_sample5 =  spark.sql(
"SELECT \
    customers.customerId, \
    avg(customers.PostalCode) as Average\
    FROM customers\
    JOIN invoices on invoices.CustomerId = customers.CustomerId\
    Group By customers.customerId \
    Having Average > 5600")
```

which customers buy more than average amount
-- having
-- inner join
-- subquery in Having and where clause

```python
select
    customer.customerId 
    format(avg(Total),2) as Average
    from customer
    join invoice on invoice.CustomerId = customer.CustomerId
    group by customer.customerId
    having Average > (select avg(Total) from invoices)
```

check if the unit prices in invoiceline table is the same as the ones in teh track table?

```python
select track.TrackId
track.unitPrice - invoice.UnitPrice as "Price difference: track vs. invoiceline table"
from track
join invoiceline on track.TrackId = invoiceline.TrackId
```

Do the totals in the invoice table match the quantity prices in the invoiceline table?

```python
select invoice.InvoiceId
invoice.Total - sum(invoiceline.UnitPrice * invoiceline.Quantity) as "Total sale differences: invoiceline vs. invoice table"
from invoice
join invoiceline on invoiceline.InvoiceId = invoice.InvoiceId
group by invoice.invoiceId
```

```python
from pyspark.sql.functions import *
import pyspark.sql.functions as f
sqlContext = SQLContext(sc)
```

```python
data = [
  ("2015-01-01 23:59:59", "2015-01-02 00:01:02", 1),
  ("2015-01-02 23:00:00", "2015-01-02 23:59:59", 2),
  ("2015-01-02 22:59:58", "2015-01-02 23:59:59", 3)]
```

```python
df = sqlContext.createDataFrame(data, ["start_time", "end_time", "id"])
type(df)
```

```python
df_new = df.select(
  df.start_time.cast("timestamp").alias("start_time"),
  df.end_time.cast("timestamp").alias("end_time"),
  df.id)
condition = \
  (to_date(df.start_time) == to_date(df.end_time)) & \
  (df.start_time + expr("INTERVAL 1 HOUR") >= df.end_time)
df.filter(condition).show()
```

```python
df_new = df.select(
    df.start_time.cast("timestamp").alias("start_time"),
    df.end_time.cast("timestamp").alias("end_time"),
    df.id)
condition = \
    (to_date(df.start_time) == to_date(df.end_time)) &\
    (df.start_time + expr("INTERVAL 1 HOUR") >= df.end_time)
df.filter(condition).show()
```

```python
from pyspark.sql import functions as F

df.select(F.date_format('timestamp','yyyy-MM-dd').alias('day')).groupby('day').count().show()
```
